```python
df = get_event_stats()
df.head()
```

```python

import matplotlib.pyplot as plt


fig, axs  = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

for i, metric in enumerate(['hammer_conversion','force_conversion','steal_conversion','stolen_pct']):
    ax = axs.flat[i]

    ax.hist(df[metric],color='blue')
    ax.set_xlim(0,1)
    ax.set_xlabel(metric)

    ax.set_ylabel("Frequency")

    ax.set_title("%s avg: %s"%(metric,round(df[metric].mean(),2)))

plt.tight_layout()
plt.show()
```
